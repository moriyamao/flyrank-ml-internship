## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

Out of 30,000 pages in the starter dataset, 54.2% are already trending down — this is the base rate any scoring method needs to beat. Declining pages have an average trend of -58%, meaning the drop is not subtle when it happens. Declining pages are also slightly younger on average (~236 days) than stable/up pages (~290 days), which goes against the simple assumption that "old content just gets stale" — suggesting a plain age-based rule would miss real cases, and a more careful scoring approach is worth building.

Note: this dataset's `trend_direction` label is a proxy calculated from the current window, not a true future outcome. A stronger version of this lane, later in the internship, would predict decline over a future window (e.g. prior 90 days of features -> next 30 days decline) rather than relying on this simpler bucket. I'll refine this by Week 4 if needed.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total pages:", len(df))
print(df["trend_direction"].value_counts(normalize=True).round(3) * 100)
print("Avg trend_pct for declining pages:", df[df.trend_direction == "down"]["trend_pct"].mean().round(1))
print("Avg content_age_days by trend_direction:")
print(df.groupby("trend_direction")["content_age_days"].mean().round(0))
```
